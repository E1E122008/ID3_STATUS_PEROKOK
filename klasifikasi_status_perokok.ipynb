```python
positif = iris[iris["smoking"] == 1].shape[0]
negatif = iris[iris["smoking"] == 0].shape[0]

print("Jumlah Perokok: ", positif)
print("Jumlah orang yang bukan Perokok: ", negatif)
```

```python
#memisakhan atribut dengan label dimana x itu atribut dan y itu label
X = iris[['age','height(cm)','weight(kg)']]
y = iris['smoking']


#membagi dataset menjadi data latih dan data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.1, random_state=123)
from sklearn.tree import DecisionTreeClassifier

#membuat model decision tree
tree_model= DecisionTreeClassifier()

#melatih model menggunakan data latih dengan fungsi fit
tree_model= tree_model.fit(X_train,y_train)
```

```python
#evaluasi model untuk meguji seberapa baik model yang digunakan dan akurasi

from sklearn.metrics import accuracy_score

y_pred =  tree_model.predict(X_test)

accuracy_score =round(accuracy_score(y_pred, y_test), 3)

print('Accuracy: ',accuracy_score)
```

```python
#prediksi model dengna tree model yang sdh dibuat dengan varibel (['age','height(cm)','weight(kg)'])

print(tree_model.predict([[6.2, 3.3, 2.0]])[0])
```

```python
from sklearn.tree import export_graphviz
from IPython.display import Image
import graphviz

# Mendefinisikan nama fitur dan nama kelas
feature_names = ['age', 'height(cm)', 'weight(kg)']
class_names = ['0', '1']  # Ganti dengan kelas yang sesuai dengan data Anda

# Membuat visualisasi model pohon keputusan langsung
dot_data = export_graphviz(tree_model, out_file=None,
                           feature_names=feature_names,
                           class_names=class_names,
                           rounded=True,
                           filled=True)

graph = graphviz.Source(dot_data)
graph
```
